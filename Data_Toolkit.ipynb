1. What is NumPy, and why is it widely used in Python?

    NumPy is a Python library used to work with nd-array and perform high-performance numerical operation.
    
    It is widely used in Python because:
    
    ->Fast Performance
    
    ->Efficient Memory Storage

2. How does broadcasting work in NumPy?

    Broadcasting in NumPy allows arithmetic operations on arrays of different shapes by automatically expanding the smaller array to match the larger one. Two dimensions are compatible if they are equal or one of them is 1.

3.  What is a Pandas DataFrame?

    A Pandas DataFrame is a 2-dimensional, labeled data structure
    used to store and manipulate data in rows and columns.

4.  Explain the use of the groupby() method in Pandas.

    The use of groupby() method in Pandas are:

    ->Find average, sum, count, max, min per group

    ->Analyze department-wise, class-wise, category-wise data

    -> Perform aggregation on grouped data

    -> Apply custom functions on each group

5.  Why is Seaborn preferred for statistical visualizations?

    Seaborn is preferred for statistical visualization because it makes statistical plots easy, attractive, and DataFrame-friendly.

6.  What are the differences between NumPy arrays and Python lists?

    NumPy arrays are faster and more memory-efficient than Python lists. They store homogeneous data and support vectorized operations and broadcasting, whereas Python lists can store heterogeneous data and require loops for numerical computations.
    Additionally.python lists use more memory compared to NumPy array.

8.  What does the term “vectorized operation” mean in NumPy?

    A vectorized operation in NumPy means performing an operation on entire arrays (vectors) at once, instead of using Python loops to process elements one by one.

9. How does Matplotlib differ from Plotly?

15. How does NumPy handle multidimensional arrays?

    NumPy handles multidimensional arrays using its ndarray object, which can store data in 1D, 2D, 3D, or nD structures. It also allows efficient element-wise operations, indexing, slicing, and reshaping across multiple dimensions.

16. What is the role of Bokeh in data visualization?

    Bokeh is a Python library for interactive and web-based data visualizations.

17. Explain the difference between apply() and map() in Pandas.

    map() :

    ->Used only with Pandas Series.

    ->Applies a function, dictionary, or mapping element-wise to each item in a Series.

    ->Cannot be used directly on a DataFrame.

    apply() :

    ->Works on both Series and DataFrames.

    ->On a Series, behaves like map().

    ->On a DataFrame, it can apply a function row-wise or column-wise.

18. What are some advanced features of NumPy?

    Advanced features of NumPy include support for multidimensional arrays, broadcasting, vectorized operations, linear algebra functions, statistical computations, random number generation, boolean indexing, and efficient file I/O, all optimized for speed and memory.

19. How does Pandas simplify time series analysis?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
#1. How do you create a 2D NumPy array and calculate the sum of each row?

arr = np.array([[1, 22, 43], [4, 15, 26], [72, 28, 9]])

rowSum = np.sum(arr, axis=1)

print(rowSum)
```

```python
#2. Write a Pandas script to find the mean of a specific column in a DataFrame

data = {'Name' : ["Jihyo","Jeongyan","Nayeon","Tyuzu","Momo","Dahyun","Sana","Mina","Chaeyoung"],'Age' : [28,29,30,26,29,26,29,28,26]}

df = pd.DataFrame(data)

meanAge = df['Age'].mean()

print(f"Mean of Twice member's age is {meanAge}")
```

```python
#3. Create a scatter plot using Matplotlib

x =[2,3,5,1,6]
y =[3,1,4,5,2]

plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
```

```python
#4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

Kpop_band = {'Year_of_debut' : [2013,2015,2016,2018,2019],'No_of_members' : [7,13,9,5,5],'Popularity' : [10,6,8,4,7 ]}
df = pd.DataFrame(Kpop_band)

correlation_matrix = df.corr()
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#5. Generate a bar plot using Plotly.

Bangtan = {'Album' : ['Map of Soul_7','Love Yourself_Answer','BE','Proof','Butter'],'Sales' : [5000000,4000000,3000000,2000000,7000000]}
df = pd.DataFrame(Bangtan)

fig = px.bar(df, x='Album', y='Sales', title='Sales of Album')
fig.show()
```

```python
#6. Create a DataFrame and add a new column based on an existing column.

df = pd.DataFrame({'Name' : ['Yeonjun','Beomgyu','Taehyun','Soobin','Kai'],'Age' : [26,24,23,25,22]})

print(df)

df['Member'] = df['Age'].apply(lambda x:'Hyung' if x>23 else 'Maknae' )
print("-----------------------------")
print(df)
```

```python
#7. Write a program to perform element-wise multiplication of two NumPy arrays.

a = np.array([22,42,11,5,23])
b = np.array([32,11,45,16,37])

result = a*b

print("Array a",a)
print("Array b",b)
print("Element wise multiplication",result)
```

```python
#8. Create a line plot with multiple lines using Matplotlib.

x = [2,3,6,1,7,4]
y1 = [3,1,5,7,9,2]
y2 = [1,4,2,9,3,5]
y3 = [2,3,5,6,1,4]

plt.plot(x,y1,label='Line 1',color = 'blue')
plt.plot(x,y2,label='Line 2',color = 'purple')
plt.plot(x,y3,label='Line 3',color = 'green')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')
plt.legend()
plt.show()
```

```python
#9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold

actors = ({'Name' : ['Wooshik','Hyunsik','Bogum','Seojoon','Dohyun'],'Experience' : [12,11,13,15,10]})
df = pd.DataFrame(actors)

threshold = 11

filtered_df = df[df['Experience'] > threshold]

print(filtered_df)
```

```python
#10. Create a histogram using Seaborn to visualize a distribution.

data = [12,16,20,7,14,15,4,9,21,19,5,10]

sns.histplot(data,bins=5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Distribution')
plt.show()
```

```python
#11. Perform matrix multiplication using NumPy.

a = np.array([[5,7],
             [4,6,]])
b = np.array([[11,5],
             [2,11]])
m = np.dot(a,b.T)

print("Matrix a\n", a)
print("Matrix b\n", b)
print("Matrix multiplication\n", m)
```
